```python
df = movie_rating_df.pivot_table(columns= 'title', index='user_id', 
                            values='rating', aggfunc='mean')
```

```python
# 타이타닉 영화와, 전체 영화의 상관계수를 구하라.

# 즉, 특정 컬럼과 전체 컬럼의 상관계수 구하는 방법
```

```python
df['Titanic (1997)']
```

```python
# 타이타닉과 모든영화의 상관계수 측정
corr_titanic = df.corrwith(  df['Titanic (1997)']  )
```

```python
corr_titanic = corr_titanic.to_frame()
```

```python
corr_titanic.columns = ['correlation']
```

```python
corr_titanic
```

```python
rating_mean_count_df['count']
```

```python
corr_titanic = corr_titanic.join(  rating_mean_count_df['count']  )
```

```python
corr_titanic
```

```python
corr_titanic.dropna(inplace = True)
```

```python
corr_titanic.sort_values('correlation', ascending=False)
```

```python
# 위의 방법은 리뷰갯수 1개짜리도 포함되어있으니까 신뢰도가 떨어진다.
# 따라서 리뷰갯수가 80개이상인 데이터를 가지고 상관계수를 보자.
```

```python
recom_movies.columns = ['correlation']
```

```python
# 추천된 영화를 저장하기 위한 비어있는 데이터프레임이 필요하고
# 반복문을 이용해서, 내가 별점 준 영화를 자동으로 처리하도록 만들자
```

```python
my_rating.shape[0]
```

```python
similar_movie_list = pd.DataFrame()
for index in range( my_rating.shape[0] ) :
  movie_title = my_rating['Movie Name'][index]
  recom_movies = corr_movie[movie_title].dropna().sort_values(ascending=False).to_frame()
  recom_movies.columns = ['correlation']
  recom_movies['weight'] = my_rating['Ratings'][index] * recom_movies['correlation']
  similar_movie_list = similar_movie_list.append(recom_movies)

```

```python
similar_movie_list
```

```python
# 1. weight 로 정렬한다.
```

```python
similar_movie_list.sort_values('weight', ascending = False, inplace=True)
```

```python
# 2. 내가 본 영화는, 삭제한다.
```

```python
my_rating
```

```python
similar_movie_list.drop('101 Dalmatians (1996)' , axis = 0)
```

```python
for name in my_rating['Movie Name'] :
  if name in similar_movie_list.index :
    similar_movie_list.drop( name  , axis = 0, inplace=True)
```
