```python
import pandas as pd
```

```python
movie_df = pd.read_json('movie.json')
```

```python
rating_df = pd.read_json('rating.json')
```

```python
movie_df.head(2)
```

```python
rating_df.head(2)
```

```python
movie_df.rename(columns={'id' : 'movieId'}, inplace=True)
```

```python
movie_df
```

```python
movie_rating_df = pd.merge(movie_df, rating_df, on = 'movieId', how='left')
```

```python
movie_rating_df
```

```python
df = movie_rating_df.pivot_table(index='userId', columns='title', values='rating')
```

```python
df
```

```python
corr_movie = df.corr(min_periods= 20)
```

```python
corr_movie
```

```python
# 특정 유저의 맞춤형 영화 추천!
# userId 1 번인 사람의 예
```

```python
user_3 = rating_df.loc[  rating_df['userId'] == 3 , ]
```

```python
user_3_df = pd.merge( user_3 , movie_df, on = 'movieId')[['title','rating']]
```

```python
user_3_df
```

```python
user_3_df['title'][0]
```

```python
recom_movies = corr_movie['Killing Room, The'].dropna().sort_values(ascending=False).to_frame()
```

```python
recom_movies.columns = ['correlation']
```

```python
recom_movies['weight'] = recom_movies['correlation'] * user_3_df['rating'][0]
```

```python
recom_movies
```

```python
similar_movies_list = pd.DataFrame()

for i in range( user_3_df.shape[0] ) :
  movie_title = user_3_df['title'][i]
  recom_movies = corr_movie[movie_title].dropna().sort_values(ascending=False).to_frame()
  recom_movies.columns = ['correlation']
  recom_movies['weight'] = recom_movies['correlation'] * user_3_df['rating'][i]
  similar_movies_list = pd.concat(  [similar_movies_list, recom_movies] )
```

```python
similar_movies_list.sort_values('weight', ascending = False, inplace=True)
```

```python
for name in user_3_df['title'] :
  if name in similar_movies_list.index :
    similar_movies_list.drop(name, axis = 0, inplace=True)
```

```python
similar_movies_list.reset_index()['title'].value_counts()
```

```python
recom_df = similar_movies_list.groupby('title')['weight'].max().sort_values(ascending=False).head(10).to_frame().reset_index()
# 중복추천된 영화는 weight가 높은거로 나타나게 해라.
```

```python
recom_df = pd.merge( recom_df , movie_df, on = 'title')
```
